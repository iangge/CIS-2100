# Order of operations
<ul><li>Exponentiation has the highest precedence.
    <li> multiplication and division have higher priority than addition and subtraction.
    <li>Operations with similar precedence are evaluated from left to right.
    <li> Parantheses change the precedence. Anything inside parantheses will be calculated first.

```python
# first 2**3=8 then 2+8=10
2 + 2**3
```

```python
#first 3*4=12 then 2+12=14
2+3*4
```
